```python
# create a file and write to it

with open('./text.txt', 'w') as file:
  file.write("Hello from Coderslang")
```

```python
# append the file content
name = input("Enter the lang name: ")
with open('./text.txt', 'a') as file:
  file.write(f"\nHello from {name}")
```

```python
# read the file
with open('./text.txt', 'r') as file:
  # option 1: read entire content
  readContent = file.read()
  print(readContent)

with open('./text.txt', 'r') as file:
  # option 2: read one line
  readContentline = file.readline()
  print(readContentline, end='')  # end='' to avoid double new line

with open('./text.txt', 'r') as file:
  # option 3: read all lines as a list
  readContentlines = file.readlines()
  print(readContentlines)
```

```python
# use os.path.exists() to check if a file exists before opening it
import os
filename = 'text.txt'
if os.path.exists(filename):
  with open(filename, 'r') as file:
    content = file.read()
    print(content)
else:
  print(f"{filename} does not exist.")
  print("Please check the file path and try again.")
  print("Exiting program.")
  
```

```python
# rb (read binary) mode
with open('./text.txt', 'rb') as file:
  content = file.read()
  print(content)
```

```python
# find the file with extension .html and read the html file
import os
for file in os.listdir('.'):
  if file.endswith('.html'):
    with open(file, 'r') as f:
      html_content = f.read()
      print(html_content)
```

```python
# Iterate through all files in the current directory, find those ending with ".html",
# open each one in text mode, then print it line-by-line with leading/trailing whitespace removed.
import os
for file in os.listdir('.'):
  if file.endswith('.html'):
    with open(file, 'r') as f:
      for line in f:
        print(line.strip())
```

```python
# magic numbers to find file type
file_path = './IMG_7576.JPG'

def get_file_type(file_path):
  with open(file_path, 'rb') as file:
    file_signature = file.read(16)

    if file_signature.startswith(b'\x89PNG\r\n\x1a\n'):
      return 'PNG Image'
    if file_signature.startswith(b'\xFF\xD8\xFF'):
      return 'JPEG Image'
    if file_signature.startswith(b'GIF87a') or file_signature.startswith(b'GIF89a'):
      return 'GIF Image'
    # Text files usually do NOT have a fixed magic header.
    # Treat as text if the bytes are decodable as UTF-8 (including ASCII).
    try:
      file_signature.decode('utf-8')
      return 'Text File'
    except UnicodeDecodeError:
      return 'Unknown file type'

file_type = get_file_type(file_path)
print(f"The file type of {file_path} is: {file_type}")
```
