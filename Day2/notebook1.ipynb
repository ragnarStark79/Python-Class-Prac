```python
# check the leap year

def is_leap_year(year: int) -> bool:
  match year:
    case y if y % 400 == 0:
      return True
    case y if y % 100 == 0:
      return False
    case y if y % 4 == 0:
      return True
    case _:
      return False

year = int(input("Enter a year: "))
if is_leap_year(year):
  print(f"{year} is a leap year.")
else:
  print(f"{year} is not a leap year.")
```

```python
# check if triangle is valid
def is_valid_triangle(a: float, b: float, c: float) -> str:
    if a + b > c and a + c > b and b + c > a:
        return "Valid"
    else:
        return "Invalid"
      
side_a = float(input("Enter side a: "))
side_b = float(input("Enter side b: "))
side_c = float(input("Enter side c: "))
result = is_valid_triangle(side_a, side_b, side_c)
print(f"The given triangle is {result}")
```

```python
# check the right angle triangle
def is_valid_right_angle_triangle(a: float, b: float, c: float) -> str:
    if a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2:
        return "Valid"
    else:
        return "Invalid"
      
side_a = float(input("Enter side a: "))
side_b = float(input("Enter side b: "))
side_c = float(input("Enter side c: "))
result = is_valid_right_angle_triangle(side_a, side_b, side_c)
print(f"The given right-angle triangle is {result}")
```

```python
# check equilateral, isosceles, scalene triangle
def triangle_type(a: float, b: float, c: float) -> str:
  match (a, b, c):
    case (x, y, z) if x == y == z:
      return "Equilateral Triangle"
    case (x, y, z) if x == y or y == z or x == z:
      return "Isosceles Triangle"
    case _:
      return "Scalene Triangle"
side_a = float(input("Enter side a: "))
side_b = float(input("Enter side b: "))
side_c = float(input("Enter side c: "))
result = triangle_type(side_a, side_b, side_c)
print(f"The given triangle is {result}")
```
