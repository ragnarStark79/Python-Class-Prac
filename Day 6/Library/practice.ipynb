```python
# list, dictionary, tuple and set practices
list1 = ['alice', 'rocky', 'local', 'harry', 'dobby', 'snape', 'voldemort', 'harry', 'hermione', 'ron', 'dumbledore', 'sirius']
print("\n".join(m for m in dir(list1) if not (m.startswith('__') and m.endswith('__'))))

```

```python
# append to list1
list1.append('raj')
print(f"Appended: {list1}")
```

```python
# add at index
list1.insert(0, 'alex')
print(f"Inseted: {list1}")
```

```python
# replace at index
list1[0] = 'alina'
print(f"Replaced: {list1}")
```

```python
# index of an element
index_of_ron = list1.index('ron')
print(f"Index of 'ron': {index_of_ron}")
```

```python
# remove an element (by value)
list1.remove('local')
print(f"Removed 'local': {list1}")
```

```python
# pop an element (last element by default otherwise index can be provided)
popped_element = list1.pop()
print(f"Popped element: {popped_element}, List after pop: {list1}")
```

```python
# delete the element
del list1[0]
print(f"Deleted element at index 0: {list1}")

# delete a slice
del list1[1:3]
print(f"Deleted elements from index 1 to 2: {list1}")

# delete the entire list
# del list1
# print("Deleted the entire list.")
```

```python
# reverse the list
list1.reverse()
print(f"Reversed list: {list1}")
list1.reverse()  # reversing back to original for further operations
print(f"Reversed back to original: {list1}")
```

```python
# coount occurrences of an element
count_harry = list1.count('harry')
print(f"Count of 'harry': {count_harry}")
```

```python
# copy the list
list2 = list1.copy()
print(f"Copied list2 from list1: {list2}")
```

```python
# expand the list by another list
list3 = ['luna', 'neville']
list1.extend(list3)
print(f"Extended list1 by list3: {list1}")
```

```python
# clear the list
list1.clear()
print(f"Cleared list1: {list1}")
```
