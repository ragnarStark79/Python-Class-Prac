```python
# tuple
tuple1 = (1, 2, 3)
tuple2 = ('a', 'b', 'c')
methods = dir(tuple1)

# filter out dunder methods
methods = [m for m in methods if not (m.startswith("__") and m.endswith("__"))]

print("Tuple methods:")
for method in range(len(methods)):
    print(f"{method + 1}. {methods[method]}")
```

```python
# maximum value
max_value = max(tuple1)
print(f"Maximum value in the tuple: {max_value}")

# minimum value
min_value = min(tuple1)
print(f"Minimum value in the tuple: {min_value}")

# reverse the tuple
reversed_tuple = tuple1[::-1]
tuple2 = ('a', 'b', 'c')
reversed_tuple2 = tuple(reversed(tuple2))

print(f"Reversed tuple: {reversed_tuple}")
print(f"Reversed tuple2: {reversed_tuple2}")
```
