```python
# list1 as key and list2 as value
list1 = ["name", "age", "city"]
list2 = ["Alice", 30, "New York"]
my_dict2 = dict(zip(list1, list2))
print(my_dict2)
```

```python
# convert my_dict to list of tuples
list_of_tuples = list(my_dict.items())
print(list_of_tuples)

# convert my_dict keys into list
print(list(my_dict.keys()))

# convert my_dict into tuple of tuple
print(tuple(my_dict.items()))
```

```python
name = ["alice", "gg"]
age = ["22", "33"]

dicv = dict(list(zip(name, age)))
print(dicv)
```

```python
# update existing key value
name = ["alice", "gg"]
age = ["22", "33"]


dicv = dict(zip(name, age))
print(dicv)
dicv.update({"alice": "25"})
print(dicv)
dicv["alice"] = "21"
print(dicv)
```
