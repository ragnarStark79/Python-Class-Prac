```python
my_set = {2, 3, 4, 5}
print(my_set)

print("\n".join(m for m in dir(my_set) if not (m.startswith('__') and m.endswith('__'))))
```

```python
# update set with multiple values
my_set = {2, 3, 4, 5}
print(my_set)

# update set with list
my_set.update([6, 7])
print(my_set)

# add single value to set
my_set.add(8)
print(my_set)

# update existing value in set by converting to list and back
temp_list = list(my_set)
temp_list[0] = 1
my_set = set(temp_list)
print(my_set)
```
