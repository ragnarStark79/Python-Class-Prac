```python
# list
name = ['Alice', 'Bob', 'Charlie']
for n in name:
  print([n, n.upper(), n.lower()])
```

```python
n = dir(name)

for i in n:
  for j in i:
    print(j, end='')
  print()
```

```python
# list
a = [1, 2, 3, 4, 5]
print(id(a))

# append 
a.append(6)
print((a))

# insert
a.insert(0, 0)
print((a))

# reverse
a.reverse()
print((a))

a.reverse()
print((a))

# ORG
a.extend([7, 8, 9])
print((a)) 
```

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]

print(list1 + list2)
list3 = list1.__add__(list2)
print(list3)

# add list2 first then list1
list4 = list2.__add__(list1)
print(list4)

# insert (use to insert at a specific index)
list1.insert(1, 10)
print(list1)
```

```python
# list ahving all the methods of list1 then print it one by one
methods = dir(list1)
# count = 0
# for method in methods:
#     count += 1
#     print(f"{count}. {method}")
    
# remove dunder methods
methods = [m for m in methods if not (m.startswith("__") and m.endswith("__"))]

for i in range(len(methods)):
    print(f"{i + 1}. {methods[i]}")
```
