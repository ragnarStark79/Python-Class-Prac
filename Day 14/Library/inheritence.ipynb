```python
class User:
    def __init__(self, name):
        self.name = name
        self.gender = 'male'
        
    def login(self):
        print(f"{self.name} logged in.")
        
# child 
class Student(User):
  def __init__(self, name):
    super().__init__(name)
      
  def enroll(self):
    print(f"{self.name} Enrolled in the course.")
        
  def login(self):
    print (f"Student logged in.")
    super().login()
    print("Super Student logged in")
        
        
student = Student('Ragnar')
student.enroll()
student.login()

print(student.name)
```

```python
class Parent:
    def __init__(self, number):
        self.__number = number

    def get_number(self):
        return self.__number


class Child(Parent):
    def __init__(self, val, number):
        super().__init__(number)
        self.__val = val

    def get_val(self):
        return self.__val

    def get_parent_number(self):
        return self.get_number()


child = Child(10, 20)
print("Parent number:", child.get_parent_number())
print("Child value:", child.get_val())
```

```python
# multilevel classes
class GrandParent:
    def __init__(self, gp_val):
        self.__gp_val = gp_val

    def get_gp_val(self):
        return self.__gp_val
class Parent(GrandParent):

    def __init__(self, p_val, gp_val):
        super().__init__(gp_val)
        self.__p_val = p_val
    def get_p_val(self):
        return self.__p_val
class Child(Parent):

    def __init__(self, c_val, p_val, gp_val):
        super().__init__(p_val, gp_val)
        self.__c_val = c_val

    def get_c_val(self):
        return self.__c_val

child = Child(10, 20, 30)
print("Child value:", child.get_c_val())
print("Parent value:", child.get_p_val())
print("GrandParent value:", child.get_gp_val())
```

```python
class A:
  def m1(self):
    return 20


class B(A):
  def m1(self):
    val = super().m1() + 20
    return val


class C(B):
  def m1(self):
    # ERROR: this calls C.m1() again (recursion) instead of calling the parent method.
    # It will recurse indefinitely and raise: RecursionError: maximum recursion depth exceeded
    # Fix would be: val = super().m1() + 20
    val = super().m1() + 20
    return val


obj = C()
print(obj.m1())
```

```python
# abstraction classes: It's a way to enforce certain methods to be implemented in child classes.
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):
    def start(self):
        print("Car started")

class Bike(Vehicle):
    def start(self):
        print("Bike started")
car = Car()
car.start()
bike = Bike()
bike.start()
```
