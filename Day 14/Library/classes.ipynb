```python
# class composition:
# In a Bank â†’ Account system, use composition

class Bank:
    def __init__(self, name):
        self.name = name
        self.accounts = []

    def add_account(self, account):
        self.accounts.append(account)

    def get_accounts(self):
        return self.accounts
class Account:
    def __init__(self, account_number, balance=0):
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
        else:
            raise ValueError("Insufficient funds")

    def get_balance(self):
        return self.balance
      
    def __str__(self):
        return f"Account {self.account_number}: Balance {self.balance}"
# Example usage
if __name__ == "__main__":
    bank = Bank("My Bank")
    
    account1 = Account("123456", 1000)
    account2 = Account("654321", 500)

    bank.add_account(account1)
    bank.add_account(account2)

    print(f"Accounts in {bank.name}:")
    for account in bank.get_accounts():
        print(account)

    account1.deposit(200)
    print(f"After deposit, {account1}")

    account2.withdraw(100)
    print(f"After withdrawal, {account2}")
```

```python
class ValentineGift:
  def __init__(self, gift_name: str, value: float):
    self.gift_name = gift_name
    self.value = value

  def present(self) -> str:
    return f"Gift presented: {self.gift_name}"


class LoverAcceptance:
  """
  Composition (strong ownership): LoverAcceptance creates and owns ValentineGift internally.
  When LoverAcceptance is destroyed, its gift becomes unreachable.
  """
  def __init__(self, gift_name: str, value: float):
    self._gift = ValentineGift(gift_name, value)

  def accept_gift(self) -> str:
    return f"Accepted with joy! Thank you for the {self._gift.gift_name}."

  def reject_gift(self) -> str:
    return f"Rejected... The {self._gift.gift_name} can't mend this broken heart."

  def __del__(self):
    # Explicitly drop reference (helps demonstrate strong ownership in a notebook).
    self._gift = None


# Example usage / demonstration
lover = LoverAcceptance("Roses", 29.99)
print(lover._gift.present())
print(lover.accept_gift())
print(lover.reject_gift())

# Strong ownership demo: after deleting 'lover', the internal gift is no longer accessible.
del lover
try:
  print(lover._gift.present())  # noqa: F821 (lover is deleted)
except Exception as e:
  print("Gift is no longer accessible:", type(e).__name__, str(e))
```

```python
# composition class: class Hero: Help() and class heroine: Hero(Help())
class Hero:
  def __init__(self, name):
    self.name = name

  def help(self):
    return f"{self.name} is helping!"


class Heroine:
  def __init__(self, name):
    self.name = name
    self.hero = Hero("Superman")

  def call_for_help(self):
    return f"{self.name} calls for help! {self.hero.help()}"


heroine = Heroine("Wonder Woman")

print(heroine.call_for_help())
```
