```python
# replace the given string with another string
message = "Welcome to the world of Python. Python is great for AI and ML."
message = message.replace("Python", "COD LANG")

```

```python
# replace with list comprehension
messages = ["Welcome to the world of Python. Python is great for AI and ML.",
            "Python is a popular programming language.",
            "Learning Python can be fun and rewarding."]
updated_messages = [msg.replace("Python", "COD LANG") for msg in messages]
for msg in updated_messages:
    print(msg)
```

```python
# replace multiple strings using a loop
message = "Welcome to the world of Python.\nPython is great for AI and ML.\nPython is easy to learn."
replacements = {
    "Python": "COD LANG",
    "AI": "Artificial Intelligence",
    "ML": "Machine Learning"
}

for old, new in replacements.items():
    message = message.replace(old, new)
print(message)
```
