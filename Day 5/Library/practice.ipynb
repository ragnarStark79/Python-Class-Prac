```python
# check palindrome
def is_palindrome(s):
    cleaned = ''.join(c.lower() for c in s if c.isalnum())
    return cleaned == cleaned[::-1]
  
print(is_palindrome("A man, a plan, a canal: Panama"))
print(is_palindrome("Hello"))
```

```python
# extract domain from email
def extract_domain(email):
    return email.split('@')[-1]
  
print(extract_domain("admin@domain.com"))

# another way to extract domain from email
mail_address = "user@example.com"
domain = mail_address.find("@")
print(mail_address[domain + 1:])
```

```python
# extract file extension
def get_file_extension(filename):
    return filename.split('.')[-1]
  
print(get_file_extension("file.pdf"))
print(".".join("file.tar.gz".split(".")[1:]))
```

```python
# mask email address
def mask_email(email):
    local, domain = email.split('@')
    masked_local = local[:3] + "*" * (len(local) - 3)
    return masked_local + "@" + domain
  
print(mask_email("aflkasdfhlka@example.com"))
```

```python
# check if the url is http/https or not
def check_url(url):
    if url.startswith("http") or url.startswith("https"): return "Valid URL"
    return "Invalid URL"

print(check_url("https://www.codelang.org"))
print(check_url("htps://www.codelang.org"))
```

```python
# replace multiple spaces with a single space
def replace_multiple_spaces(sentence):
    return ' '.join(sentence.split())

sentence = "This is a   test sentence."
print(replace_multiple_spaces(sentence))
```

```python
# check password strength

def check_password(password: str) -> str:
    has_length = len(password) >= 8
    for l in password:
        has_lowercase = any(l.islower() for l in password)
    has_uppercase = any(l.isupper() for l in password)
    has_number = any(l.isdigit() for l in password)
    print(has_length, has_lowercase, has_uppercase, has_number)


    if has_length and has_lowercase and has_uppercase and has_number:
        return "Strong Password."
    else:
        return "Weak Password."


password = "asdfkAAsfddG7657JHJHJjsdF"
print(check_password(password))
```
