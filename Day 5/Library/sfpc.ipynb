```python
import re

def find_word_regex(string, word):
    # re.IGNORECASE makes it find both 'To' and 'to'
    # .start() gives the index of the match
    indices = [m.start() for m in re.finditer(word, string, re.IGNORECASE)]
    return f"'{word}' found at indexes: {indices}"

string = "Welcome to the world of titan. Tighten your seatbelts to enjoy this world further."
print(find_word_regex(string, "to"))
```
