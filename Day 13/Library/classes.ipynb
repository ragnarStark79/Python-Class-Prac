```python
# Simple class

class User:
    def __init__(self, name, email):
        # instance variables
        self.name = name
        self.email = email
        
    def display_info(self):
        print(f"Name: {self.name}, Email: {self.email}")
        
# creating an instance of the User class
user1 = User("Alice", "alice@example.com")
user1.display_info()
```

```python
# class with private and public variables
class User:
    def __init__(self, name, email):
        # public instance variables
        self.name = name
        self.email = email
        # private instance variable
        self.__password = "secret"
        
    def display_info(self):
        print(f"Name: {self.name}, Email: {self.email}")
        
    def set_password(self, password):
        self.__password = password
        
    def get_password(self):
        return self.__password

# creating an instance of the User class
user1 = User("Alice", "alice@example.com")
user1.display_info()
print(user1.get_password())

# change the password
# private variable cannot be accessed directly
# print(user1.__password)  # This will raise an AttributeError because __password is a private variable and cannot be accessed directly
# print(user1.__password)  # This will raise an AttributeError
# private variable can't be modified directly but can be modified using a public method for example set_password
user1.set_password("new_secret")
print(user1.get_password())


# This will still print "new_secret" because __password is a private variable and cannot be modified directly
user1.__password = "hacked"
print(user1.get_password())

  
# modifying the private variable using the name mangling syntax
user1._User__password = "hacked"
print(user1.get_password())  # This will print "hacked" because we have modified the private variable using the name mangling syntax
```

```python
class Bank:
  def __init__(self, owner, balance=0):
    self.owner = owner
    self.__balance = balance
    
  def __validate_amount(self, amount):
    if amount <= 0:
      raise ValueError("Amount should be greater than 0")
    return True
  
  def deposit(self, amount):
    if self.__validate_amount(amount):
      self.__balance += amount
      print(f"{self.owner} deposited amount: ${amount} >>> New Balance: ${self.__balance}")
      
  def withdraw(self, amount):
    if self.__validate_amount(amount):
      if amount < self.__balance:
        self.__balance -= amount
        print(f"{self.owner} withdraw amount: ${amount} >>> New Balance: ${self.__balance}")
      else:
        print("Insufficient amount")
    else:
      print("Insufficient amount or Invalid amount")

  def display(self):
    print(f"Total available amount: {self.__balance}")
      
bank = Bank('Ragnar', 100000)
bank.deposit(999)
bank.withdraw(999)
bank.display()
      
```

```python
class Bank:
  def __init__(self, owner, balance=0):
    self.owner = owner
    self.__balance = balance
    
  def __validate_amount(self, amount: float) -> None:
    if amount <= 0:
      raise ValueError("Amount should be greater than 0")
    return True
  
  def deposit(self, amount: float) -> None:
    if self.__validate_amount(amount):
      self.__balance += amount
      print(f"           {self.owner}\ndeposited amount >>> New Balance: \n  ${amount}               ${self.__balance}")
      
  def withdraw(self, amount: float) -> None:
    if self.__validate_amount(amount):
      if amount < self.__balance:
        self.__balance -= amount
        print(f"           {self.owner}\nwithdrew amount >>> New Balance: \n  ${amount}               ${self.__balance}")
      else:
        print("Insufficient amount")
    else:
      print("Insufficient amount or Invalid amount")

  def display(self) -> None:
    print(f"    Total available Balance \n           ${self.__balance}")
      
bank = Bank('Ragnar', 100000)
bank.deposit(999)
bank.withdraw(999)
bank.display()
      
```
