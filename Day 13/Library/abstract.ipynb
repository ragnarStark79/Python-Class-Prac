```python
from abc import ABC, abstractmethod

class Thor_hammer(ABC):
    @abstractmethod
    def worthey(self):
        pass
      
    def fly(self):
      print("The hammer is flying!")
      
    def thunder(self):
      print("Thunder strikes!")
      
class Thor(Thor_hammer):
  def worthey(self):
    print("Thor is worthy of the hammer!")

thor = Thor()
thor.worthey()
thor.fly()
thor.thunder()
```

```python
class Animal:
  @abstractmethod
  def make_sound(self):
    pass
  
class Dog(Animal):
  def make_sound(self):
    print("Woof!")
    
class Cat(Animal):
  def make_sound(self):
    print("Meow!")
    
dog = Dog()
cat = Cat()
dog.make_sound()
cat.make_sound()
```

```python
class SecureLocker:
  def __init__(self, pin):
    self.__pin = pin
    self.__is_locked = True
  
  def __verify_pin(self, input_pin):
    return input_pin == self.__pin
  
  def unlock(self, input_pin):
    if self.__verify_pin(input_pin):
      self.__is_locked = False
      print("Locker unlocked!")
    else:
      print("Incorrect PIN. Locker remains locked.")
      
  def lock(self):
    self.__is_locked = True
    print("Locker locked!")
    
  def status(self):
    if self.__is_locked:
      print("Locker is locked.")
    else:
      print("Locker is unlocked.")
        
locker = SecureLocker(1234)
locker.status()
locker.unlock(1234)
locker.status()
locker.lock()
locker.status()
  
```

```python
# composition class
# In a Bank â†’ Account system, use composition

class Bank:
    def __init__(self):
        self.accounts = []

    def add_account(self, account):
        self.accounts.append(account)
        
    def show_accounts(self):
        for account in self.accounts:
            print(f"Account Holder: {account.holder_name}, Balance: {account.balance}") 
            
class Account:
    def __init__(self, holder_name, balance=0):
        self.holder_name = holder_name
        self.balance = balance
        
    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited {amount}. New balance: {self.balance}")
        
    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance: {self.balance}")
        else:
            print("Insufficient funds.")

bank = Bank()
account1 = Account("Alice", 1000)
account2 = Account("Bob", 500)
bank.add_account(account1)
bank.add_account(account2)
bank.show_accounts()
account1.deposit(200)
account2.withdraw(100)
bank.show_accounts()

  
```
