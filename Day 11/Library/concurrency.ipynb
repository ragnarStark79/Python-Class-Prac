```python
# synchronous 
import time

def task(name):
    start = time.time()
    print(f"{name} started at", start)

    time.sleep(3)

    end = time.time()
    print(f"{name} ended at", end)
    print(f"{name} took {end - start:.2f} seconds\n")


overall_start = time.time()

task("Task 1")
task("Task 2")

overall_end = time.time()
print("Total time:", overall_end - overall_start)
```

```python
# simple asynchronous example using asyncio
# NOTE: In Jupyter, don't use asyncio.run(); the notebook already has a running event loop.
import asyncio

async def task():
    await asyncio.sleep(2)
    print("Done")

async def main():
    await task()
    print("Next line")


await main()
```

```python
# simple synchronous example
import time

def task():
    time.sleep(2)
    print("Done")

task()
print("Next line")
```

```python
# Asynchronous version
import asyncio
import time

async def task(name):
    start = time.time()
    print(f"{name} started at", start)

    await asyncio.sleep(3)

    end = time.time()
    print(f"{name} ended at", end)
    print(f"{name} took {end - start:.2f} seconds\n")


async def main():
    overall_start = time.time()

    await asyncio.gather(
        task("Task 1"),
        task("Task 2")
    )

    overall_end = time.time()
    print("Total time:", overall_end - overall_start)



await main()

```
