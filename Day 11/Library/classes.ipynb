```python
# class

class Car:
  pass

car = Car()
car.brand = "Toyota"
car.color = "Red"

car1 = Car()
car1.brand = "Honda"
car1.color = "Blue"

print(car.brand)
print(car1.brand)


# __dict__ is a special attribute that stores the attributes of an object in a dictionary
print(car.__dict__)


# __dir__ is a special method that returns a list of the attributes and methods of an object
print(car.__dir__())

# __dir__ used with class returns a list of the attributes and methods of the class, including the special methods
print(Car.__dict__)

# use __builtins__.dir since `dir` is a str variable in this notebook. __builtins__ is a module that contains the built-in functions and variables of Python. dir is a built-in function that returns a list of the attributes and methods of an object. Since we have a variable named dir in this notebook, we need to use __builtins__.dir to access the built-in function.
for name in __builtins__.dir(car):
  print(name)
```

```python
class Car:
  # self is a reference to the current instance of the class
  # self is default name, you can use any name you want
  def brake(self):
    print("Car is braking")
    
  def speed(self):
    print("Car is speeding")
    
car = Car()
car.brake()
car.speed()
```

```python
class SuperHero:
  # __init__ is a special method that is called when an object is created. It is used to initialize the attributes of the object.
  def __init__(self, name, power):
    self.name = name
    self.power = power
    
  def display(self):
    print(f"{self.name} has the power of {self.power}")
    
superhero = SuperHero("Aren Yeager", "Titan Shifting")
superhero1 = SuperHero("Ironman", "Genius-level intellect and advanced technology")
superhero.display()
superhero1.display()

print(superhero)
print(superhero1)

# both superhero and superhero1 now point to same object in memory
superhero = superhero1
print(superhero)
print(superhero1)
```

```python
# class dog, attribute set_name, method bark, create 2 dog

class Dog():
  def __init__(self, name):
    self.name = name
    
    
  def set_name(self, name):
    self.name = name

  def bark(self):
    print(f"{self.name} says woof!")
    
dog1 = Dog("Buddy")
dog2 = Dog("Charlie")
dog1.bark()
dog2.bark()

dog1.set_name("Max")
dog1.bark()
```

```python
# Student Management System
class Student:
  def __init__(self, name, marks, roll_number):
    self.name = name
    self.marks = marks
    self.roll_number = roll_number
    
  def add_marks(self, marks):
    self.marks += marks
    
  def is_passed(self):
    return self.marks > 40
      
  def display(self):
    print(f"Name: {self.name}, Marks: {self.marks}, Roll Number: {self.roll_number}")
    print(f"Passed: {self.is_passed()}")

student1 = Student("Alice", 85, 101)
student2 = Student("Bob", 40, 102)
student1.display()
student2.display()

student1.add_marks(5)
student1.display()

result = student1.is_passed()
result2 = student2.is_passed()
print(f"Student 1 Passed: {result}")
print(f"Student 2 Passed: {result2}")

print(student1.__dict__)
print(student2.__dict__)
```
